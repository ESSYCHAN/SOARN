```python
#Subelement weights extracted from the model A as the predictor
subA = GRP.get_cue_map()['A'].get_finale_wts()
ele_0A = np.asarray(subA[0])
# ele_0A.shape
subx = GRP.get_cue_map()['X'].get_finale_wts()
ele_0x = np.asarray(subx[0])

subus = GRP.get_cue_map()['+'].get_finale_wts()
ele_0us = np.asarray(subus[0])
# ele_0us
```

```python
#Initial filter_maps from the conv layer 

stim_predA = A.filter_maps
stim_predA = stim_predA.flatten()

activationsA  = GRP.activate_test('A')
acti_0A = np.asarray(activationsA[0])
acti_0A.shape
stim_predA*acti_0A
############################
stim_predX = X.filter_maps
stim_predX = stim_predX.flatten()

activationsX = GRP.activate_test('X')
acti_0X = np.asarray(activationsX[0])
acti_0X.shape
stim_predX*acti_0X

######################

stim_predus = US.filter_maps
stim_predus = stim_predus.flatten()

activationsus  = GRP.activate_test('+')
acti_0us = np.asarray(activationsus[0])
acti_0us.shape
stim_predus = acti_0us*stim_predus
```

```python
# Extract weights to the US
US_images_wtsA = ele_0A[:,2, :]
US_images_wtsA.shape
# US_images

US_images_wtsX = ele_0x[:,2, :]
US_images_wtsX.shape

US_images_wtsus = ele_0us[:,2, :]
US_images_wtsus.shape
```

```python
US_images_wtsX
```

```python
# figure showing the predicted value
fig, axes = plt.subplots(4,4, figsize=(6,6))
for i in range(4): #Columns repping the trial time 
    for j in range(4): #Rows repping the convolution maps
        y = (((stim_predA* US_images_wtsA[:,i])+(stim_predX* US_images_wtsX[:,i])+(stim_predus* US_images_wtsus[:,i]))).reshape(4, 56, 56)[j]
        axes[j][i].imshow(y)
```
